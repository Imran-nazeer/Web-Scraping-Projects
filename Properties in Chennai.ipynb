```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.magicbricks.com/property-for-sale-in-tambaram-west-chennai-pppfs")
c = r.content

soup = BeautifulSoup(c, "html.parser")

all = soup.find_all("div",{"class":"flex relative clearfix m-srp-card__container"})

len(all)
```

```python
list = []
for item in all:
    dict = {}
    try:
        dict["Name of Property"]=item.find_all("div",{"class":"m-srp-card__society"})[0].find("a").text
    except:
         dict["Name of Property"]=None
    try:  
        dict["Price in Lac"]=item.find_all("div",{"class":"m-srp-card__price"})[0].text.replace("Lac","").replace(" ","").replace("₹","")
    except:
           dict["Price in Lac"]=None
    
    try:
        dict["Price/SQFT"]=item.find_all("div",{"class":"m-srp-card__area"})[0].find("span").text.replace(" ","").replace("₹","")
    except:
        dict["Price/SQFT"]=None
    
    try:
        dict["Area in SQFT"]=item.find_all("div",{"class":"m-srp-card__summary__info"})[0].text.replace("sqft","").replace("\xa0","")
    except:
         dict["Area in SQFT"]=None
            
            
    list.append(dict)
        
    
    
```

```python
import pandas as pd
df = pd.DataFrame(list)
```
